## Estrategia de resolución de los problemas

<ul>
    <li> Generé un desarrollo local a cada pregunta donde la principal diferencia entre ambos enfoques (tiempo ejecución y uso memoria), seria la forma de carga del archivo, linea por linea para memoria y mediante chunks para tiempo </li>
    <li> Se generó un parseo solo de los campos necesarios tanto para la resolución de cada pregunta, esto ahorró tanto tiempo como memoria</li>
    <li>Como estructuras de datos, se usaron diccionarios, optimizados para el conteo de hasheables (Counter) o quer facilitaran el manejo de llaves nulas (defaultdict), para encontrar top en general se uso nlargest que es un heap que en no genera overhead de memoria a diferencia del sorted (Timsort usa python) y tiene un tiempo de ejecución menor en promedio siempre y cuando el top sea bajo en cantidadde elementos, finalmente para guardar tops se uso un deque (lista linkeada doble) por su mayor rapidez para operaciones append por izquierda o derecha que lista</li>
</ul>
